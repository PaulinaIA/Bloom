### 5.2 Data Layer

Soporte para el **dataset de Marquette** (Universidad de Marquette - Planificaci√≥n Familiar Natural) y generaci√≥n de datos sint√©ticos para desarrollo.

````python
class MarquetteDataLoader:
    
    COLUMN_MAPPING = {
        'LengthofCycle': 'cycle_length',
        'LengthofMenses': 'period_length',
        'LengthofLutealPhase': 'luteal_phase',
        'EstimatedDayofOvulation': 'ovulation_day',
        'MensesScoreDayOne': 'bleeding_intensity',
        'UnusualBleeding': 'unusual_bleeding',
        'TotalNumberofPeakDays': 'peak_days',
        'TotalHighDays': 'high_fertility_days',
        'ClientAge': 'age',
        'BMI': 'bmi',
        'NumberPregnancies': 'previous_pregnancies',
        'Numberpregnancyloss': 'miscarriages',
        'CurrentlyBreastworNursing': 'breastfeeding'
    }
    
    def __init__(self, filepath: Optional[str] = None):
        self.filepath = filepath
    
    def load(self) -> pd.DataFrame:
        if self.filepath:
            return self._load_marquette()
        return self._generate_synthetic()
    
    def _load_marquette(self) -> pd.DataFrame:
        df = pd.read_csv(self.filepath)
        df = df.rename(columns={k: v for k, v in self.COLUMN_MAPPING.items() if k in df.columns})
        return self._validate(df)
    
    def _validate(self, df: pd.DataFrame) -> pd.DataFrame:
        if 'cycle_length' in df.columns:
            df = df[(df['cycle_length'] >= 21) & (df['cycle_length'] <= 45)]
        if 'ovulation_day' in df.columns:
            df = df[(df['ovulation_day'] >= 8) & (df['ovulation_day'] <= 35)]
        return df.dropna(subset=['cycle_length', 'ovulation_day'])
    
    def _generate_synthetic(self, n_users: int = 200, cycles_per_user: int = 12) -> pd.DataFrame:
        np.random.seed(42)
        data = []
        
        for user_id in range(1, n_users + 1):
            profile = self._create_profile()
            current_date = datetime(2022, 1, 1)
            
            for cycle_num in range(1, cycles_per_user + 1):
                cycle = self._create_cycle(user_id, cycle_num, current_date, profile)
                data.append(cycle)
                current_date += timedelta(days=cycle['cycle_length'])
        
        return pd.DataFrame(data)
    
    def _create_profile(self) -> Dict:
        return {
            'age': np.random.randint(20, 42),
            'regularity': np.random.uniform(0.6, 1.0),
            'base_cycle': np.clip(np.random.normal(28, 2.5), 24, 35),
            'base_luteal': np.clip(np.random.normal(13, 1.2), 10, 16),
            'bmi': np.random.normal(23, 4),
            'previous_pregnancies': np.random.choice([0, 1, 2, 3], p=[0.4, 0.3, 0.2, 0.1]),
            'miscarriages': np.random.choice([0, 1, 2], p=[0.8, 0.15, 0.05]),
            'anomaly_prone': np.random.random() < 0.15
        }
    
    def _create_cycle(self, user_id: int, cycle_num: int, start_date: datetime, p: Dict) -> Dict:
        noise = 1 - p['regularity']
        cycle_length = int(np.clip(p['base_cycle'] + np.random.normal(0, 3*noise), 21, 45))
        luteal_phase = int(np.clip(p['base_luteal'] + np.random.normal(0, 1.5*noise), 9, 17))
        ovulation_day = cycle_length - luteal_phase
        
        unusual_bleeding = 0
        if p['anomaly_prone'] and np.random.random() < 0.2:
            anomaly = np.random.choice(['bleeding', 'long', 'short_luteal'])
            if anomaly == 'bleeding':
                unusual_bleeding = 1
            elif anomaly == 'long':
                cycle_length = np.random.randint(38, 45)
                ovulation_day = cycle_length - luteal_phase
            else:
                luteal_phase = np.random.randint(8, 10)
                ovulation_day = cycle_length - luteal_phase
        
        return {
            'user_id': user_id,
            'cycle_number': cycle_num,
            'start_date': start_date,
            'cycle_length': cycle_length,
            'period_length': int(np.clip(np.random.normal(5, 1), 3, 7)),
            'ovulation_day': ovulation_day,
            'luteal_phase': luteal_phase,
            'bleeding_intensity': np.random.choice([1, 2, 3], p=[0.25, 0.5, 0.25]),
            'unusual_bleeding': unusual_bleeding,
            'peak_days': np.random.randint(1, 4),
            'high_fertility_days': np.random.randint(4, 8),
            'age': p['age'],
            'bmi': round(p['bmi'], 1),
            'previous_pregnancies': p['previous_pregnancies'],
            'miscarriages': p['miscarriages'],
            'regularity_score': round(p['regularity'], 2)
        }
````

### 5.3 Feature Engineering

````python
class BloomFeatureEngineer:
    
    CYCLE_FEATURES = [
        'cycle_length_lag1', 'cycle_length_lag2', 'cycle_length_lag3',
        'cycle_length_rolling_mean', 'cycle_length_rolling_std',
        'cycle_trend', 'personal_cycle_mean', 'personal_cycle_std', 'cycle_cv'
    ]
    
    OVULATION_FEATURES = [
        'ovulation_day_lag1', 'ovulation_day_lag2',
        'ovulation_rolling_mean', 'personal_ovulation_mean',
        'luteal_phase_lag1', 'personal_luteal_mean'
    ]
    
    HEALTH_FEATURES = ['age', 'bmi', 'previous_pregnancies', 'miscarriages']
    
    ANOMALY_FEATURES = [
        'bleeding_intensity', 'unusual_bleeding_history',
        'cycle_deviation_from_mean', 'luteal_deviation_from_mean',
        'consecutive_irregular_cycles'
    ]
    def transform(self, df: pd.DataFrame) -> pd.DataFrame:
        df = df.copy().sort_values(['user_id', 'cycle_number'])
        df = self._add_lags(df)
        df = self._add_rolling(df)
        df = self._add_personal_stats(df)
        df = self._add_anomaly_features(df)

        # 1) filtro m√≠nimo de filas (lo que ya hac√≠as)
        required = [
            "user_id", "cycle_number", "ovulation_day",
            "ovulation_day_lag1", "ovulation_day_lag2",
            "personal_ovulation_mean"
        ]
        required = [c for c in required if c in df.columns]
        df = df.dropna(subset=required)

        # 2) limpieza sobre lo que queda
        # 2.1: strings vac√≠os / espacios -> NaN
        df = df.replace(r'^\s*$', np.nan, regex=True)

        # 2.2: columnas de features que deben ser num√©ricas
        feature_cols = list(set(
            self.CYCLE_FEATURES +
            self.OVULATION_FEATURES +
            self.HEALTH_FEATURES +
            self.ANOMALY_FEATURES
        ))
        feature_cols = [c for c in feature_cols if c in df.columns]

        for c in feature_cols:
            df[c] = pd.to_numeric(df[c], errors="coerce")

        # 2.3: inf -> NaN
        df[feature_cols] = df[feature_cols].replace([np.inf, -np.inf], np.nan)

        # 2.4: quitar columnas que son TODO NaN
        non_all_nan = df[feature_cols].columns[~df[feature_cols].isna().all()].tolist()
        df = df.drop(columns=[c for c in feature_cols if c not in non_all_nan])
        feature_cols = non_all_nan  # actualizamos lista

        # 2.5: rellenar NaN restantes con mediana
        med = df[feature_cols].median(numeric_only=True)
        df[feature_cols] = df[feature_cols].fillna(med)

        # 3) asegurar que cycle_length exista para el modelo de ciclo
        if "cycle_length" in df.columns:
            df = df.dropna(subset=["cycle_length"])

        return df

    '''def transform(self, df: pd.DataFrame) -> pd.DataFrame:
        df = df.copy().sort_values(['user_id', 'cycle_number'])
        df = self._add_lags(df)
        df = self._add_rolling(df)
        df = self._add_personal_stats(df)
        df = self._add_anomaly_features(df)

        # 1) Primero: filtrar filas que tengan lo m√≠nimo para ovulaci√≥n
        required = [
            "user_id", "cycle_number", "ovulation_day",
            "ovulation_day_lag1", "ovulation_day_lag2",
            "personal_ovulation_mean"
        ]
        required = [c for c in required if c in df.columns]
        df = df.dropna(subset=required)

        # 2) Ahora, sobre lo que queda: limpieza de valores
        # 2.1: strings vac√≠os / espacios -> NaN
        df = df.replace(r'^\s*$', np.nan, regex=True)

        # 2.2: columnas de features a num√©ricas
        feature_cols = list(set(
            self.CYCLE_FEATURES +
            self.OVULATION_FEATURES +
            self.HEALTH_FEATURES +
            self.ANOMALY_FEATURES
        ))
        feature_cols = [c for c in feature_cols if c in df.columns]

        for c in feature_cols:
            df[c] = pd.to_numeric(df[c], errors="coerce")

        # 2.3: inf -> NaN y rellenar con mediana, luego 0 si alguna columna entera es NaN
        df[feature_cols] = df[feature_cols].replace([np.inf, -np.inf], np.nan)
        med = df[feature_cols].median(numeric_only=True)
        df[feature_cols] = df[feature_cols].fillna(med).fillna(0)

        # 3) Asegurar que cycle_length no sea NaN para el modelo de ciclo
        if "cycle_length" in df.columns:
            df = df.dropna(subset=["cycle_length"])

        return df'''
    # def transform(self, df: pd.DataFrame) -> pd.DataFrame:
    #     df = df.copy().sort_values(['user_id', 'cycle_number'])
    #     df = self._add_lags(df)
    #     df = self._add_rolling(df)
    #     df = self._add_personal_stats(df)
    #     df = self._add_anomaly_features(df)
    #     required = [
    #         "user_id", "cycle_number", "ovulation_day",
    #         "ovulation_day_lag1", "ovulation_day_lag2",
    #         "personal_ovulation_mean"]
    #     return df.dropna(subset=required)
    #     #return df.dropna() 
    # def transform(self, df: pd.DataFrame) -> pd.DataFrame:
    #     df = df.copy().sort_values(['user_id', 'cycle_number'])
    #     df = self._add_lags(df)
    #     df = self._add_rolling(df)
    #     df = self._add_personal_stats(df)
    #     df = self._add_anomaly_features(df)

    #     df = df.replace(r'^\s*$', np.nan, regex=True)
    #     feature_cols = list(set(self.CYCLE_FEATURES + self.OVULATION_FEATURES + self.HEALTH_FEATURES + self.ANOMALY_FEATURES))
    #     feature_cols = [c for c in feature_cols if c in df.columns]
    #     for c in feature_cols:
    #         df[c] = pd.to_numeric(df[c], errors="coerce")
    #     df[feature_cols] = df[feature_cols].replace([np.inf, -np.inf], np.nan)
    #     med = df[feature_cols].median(numeric_only=True)
    #     df[feature_cols] = df[feature_cols].fillna(med).fillna(0)
    #     required_cycle = ["cycle_length_lag1", "cycle_length_lag2", "cycle_length_lag3", "personal_cycle_mean"]
    #     required_cycle = [c for c in required_cycle if c in df.columns]
    #     df = df.dropna(subset=required_cycle + ["cycle_length"])
    #     return df

    def _add_lags(self, df: pd.DataFrame) -> pd.DataFrame:
        for lag in [1, 2, 3]:
            df[f'cycle_length_lag{lag}'] = df.groupby('user_id')['cycle_length'].shift(lag)
        for lag in [1, 2]:
            df[f'ovulation_day_lag{lag}'] = df.groupby('user_id')['ovulation_day'].shift(lag)
            df[f'luteal_phase_lag{lag}'] = df.groupby('user_id')['luteal_phase'].shift(lag)
        return df
    
    def _add_rolling(self, df: pd.DataFrame) -> pd.DataFrame:
        df['cycle_length_rolling_mean'] = df.groupby('user_id')['cycle_length'].transform(lambda x: x.shift(1).rolling(3, min_periods=1).mean())
        df['cycle_length_rolling_std'] = df.groupby('user_id')['cycle_length'].transform(lambda x: x.shift(1).rolling(3, min_periods=1).std())
        df['ovulation_rolling_mean'] = df.groupby('user_id')['ovulation_day'].transform(lambda x: x.shift(1).rolling(3, min_periods=1).mean())
        df['cycle_trend'] = df.groupby('user_id')['cycle_length'].diff()
        return df
    
    def _add_personal_stats(self, df: pd.DataFrame) -> pd.DataFrame:
        df['personal_cycle_mean'] = df.groupby('user_id')['cycle_length'].transform(lambda x: x.shift(1).expanding().mean())
        df['personal_cycle_std'] = df.groupby('user_id')['cycle_length'].transform(lambda x: x.shift(1).expanding().std())
        df['cycle_cv'] = df['personal_cycle_std'] / df['personal_cycle_mean']
        df['personal_ovulation_mean'] = df.groupby('user_id')['ovulation_day'].transform(lambda x: x.shift(1).expanding().mean())
        df['personal_luteal_mean'] = df.groupby('user_id')['luteal_phase'].transform(lambda x: x.shift(1).expanding().mean())
        return df
    
    def _add_anomaly_features(self, df: pd.DataFrame) -> pd.DataFrame:
        df['unusual_bleeding_history'] = df.groupby('user_id')['unusual_bleeding'].transform(lambda x: x.shift(1).rolling(6, min_periods=1).sum())
        df['cycle_deviation_from_mean'] = df['cycle_length'] - df['personal_cycle_mean']
        df['luteal_deviation_from_mean'] = df['luteal_phase'] - df['personal_luteal_mean']
        df['is_irregular'] = (df['cycle_length_rolling_std'] > 4).astype(int)
        df['consecutive_irregular_cycles'] = df.groupby('user_id')['is_irregular'].transform(lambda x: x.rolling(3, min_periods=1).sum())
        return df
    
    def get_feature_sets(self) -> Dict[str, List[str]]:
        return {
            'cycle': self.CYCLE_FEATURES + self.HEALTH_FEATURES,
            'ovulation': self.OVULATION_FEATURES + self.CYCLE_FEATURES[:5] + self.HEALTH_FEATURES,
            'anomaly': self.ANOMALY_FEATURES + self.CYCLE_FEATURES[:5]
        }
````

    <strong>Target:</strong> RMSE < 2 d√≠as
</div>

````python
class CycleLengthPredictor:
    
    def __init__(self):
        self.model = RandomForestRegressor(n_estimators=150, max_depth=12, min_samples_split=5,min_samples_leaf=2, random_state=42, n_jobs=-1)
        self.feature_cols: List[str] = []
        self.is_fitted = False
    
    def fit(self, X: pd.DataFrame, y: pd.Series, features: List[str]) -> 'CycleLengthPredictor':
        self.feature_cols = features
        X_clean = X[features].fillna(X[features].median())
        self.model.fit(X_clean, y)
        self.is_fitted = True
        return self
    
    def predict(self, X: pd.DataFrame) -> np.ndarray:
        X_clean = X[self.feature_cols].fillna(X[self.feature_cols].median())
        return self.model.predict(X_clean)
    
    def predict_with_interval(self, X: pd.DataFrame, confidence: float = 0.9) -> Tuple[np.ndarray, np.ndarray, np.ndarray]:
        X_clean = X[self.feature_cols].fillna(X[self.feature_cols].median())
        preds = np.array([tree.predict(X_clean) for tree in self.model.estimators_])
        alpha = (1 - confidence) / 2
        return preds.mean(axis=0), np.percentile(preds, alpha*100, axis=0), np.percentile(preds, (1-alpha)*100, axis=0)
    
    def get_feature_importance(self) -> pd.DataFrame:
        return pd.DataFrame({'feature': self.feature_cols,'importance': self.model.feature_importances_}).sort_values('importance', ascending=False)
    
    def evaluate(self, X: pd.DataFrame, y: pd.Series) -> ModelMetrics:
        y_pred = self.predict(X)
        return ModelMetrics(model_name='CycleLengthPredictor', rmse=round(np.sqrt(mean_squared_error(y, y_pred)), 3), mae=round(mean_absolute_error(y, y_pred), 3),
            r2=round(r2_score(y, y_pred), 4))
````

### 5.5 Ovulation Day Classifier

<div style="background: #fff3e0; padding: 15px; border-radius: 8px; border-left: 4px solid #ff9800;">
    <strong> Target:</strong> 95% accuracy within ¬±2 days
</div>

````python
class OvulationClassifier:
    
    def __init__(self, window_size: int = 2):
        self.window_size = window_size
        self.regressor = RandomForestRegressor(n_estimators=150, max_depth=10, random_state=42, n_jobs=-1)
        self.classifier = GradientBoostingClassifier(n_estimators=200, max_depth=6, learning_rate=0.1, random_state=42)
        self.feature_cols: List[str] = []
        self.is_fitted = False
    
    def fit(self, X: pd.DataFrame, y: pd.Series, features: List[str]) -> 'OvulationClassifier':
        self.feature_cols = features
        X_clean = X[features].fillna(X[features].median())
        self.regressor.fit(X_clean, y)
        y_binned = pd.cut(y, bins=[0, 11, 13, 15, 17, 50], labels=[0, 1, 2, 3, 4]).astype(int)
        self.classifier.fit(X_clean, y_binned)
        self.is_fitted = True
        return self
    
    def predict(self, X: pd.DataFrame) -> OvulationPrediction:
        X_clean = X[self.feature_cols].fillna(X[self.feature_cols].median())
        day_pred = int(round(self.regressor.predict(X_clean)[0]))
        confidence = max(self.classifier.predict_proba(X_clean)[0])
        return OvulationPrediction(predicted_day=day_pred, confidence=round(confidence, 3), window_start=max(8, day_pred - self.window_size), window_end=min(25, day_pred + self.window_size))
    
    def predict_batch(self, X: pd.DataFrame) -> np.ndarray:
        X_clean = X[self.feature_cols].fillna(X[self.feature_cols].median())
        return np.round(self.regressor.predict(X_clean)).astype(int)
    
    def evaluate(self, X: pd.DataFrame, y_true: pd.Series) -> Dict[str, Any]:
        y_pred = self.predict_batch(X)
        errors = np.abs(y_true - y_pred)
        return {
            'exact_accuracy': round(np.mean(errors == 0) * 100, 2),
            'within_1_day': round(np.mean(errors <= 1) * 100, 2),
            'within_2_days': round(np.mean(errors <= 2) * 100, 2),
            'mae_days': round(mean_absolute_error(y_true, y_pred), 2),
            'meets_target': np.mean(errors <= 2) >= 0.95
        }
````

### 5.6 Anomaly Detector

<div style="background: #ffebee; padding: 15px; border-radius: 8px; border-left: 4px solid #f44336;">
    <strong> Target:</strong> Recall > 90% en detecci√≥n de sangrado inusual
</div>

````python
class AnomalyDetector:
    
    def __init__(self, contamination: float = 0.1):
        self.isolation_forest = IsolationForest(contamination=contamination, random_state=42, n_estimators=100)
        self.scaler = StandardScaler()
        self.feature_cols: List[str] = []
        self.is_fitted = False
    
    def fit(self, X: pd.DataFrame, features: List[str]) -> 'AnomalyDetector':
        self.feature_cols = features
        X_clean = X[features].fillna(X[features].median())
        X_scaled = self.scaler.fit_transform(X_clean)
        self.isolation_forest.fit(X_scaled)
        self.is_fitted = True
        return self
    
    def detect(self, record: pd.Series) -> List[AnomalyAlert]:
        alerts = []
        
        if record.get('luteal_phase', 14) < 10:
            alerts.append(AnomalyAlert(
                anomaly_type=AnomalyType.SHORT_LUTEAL,
                severity='high',
                description=f"Fase l√∫tea corta ({record['luteal_phase']} d√≠as). Normal: 11-16 d√≠as.",
                recommendation="Puede dificultar implantaci√≥n. Considera consultar especialista.",
                confidence=0.95
            ))
        
        if record.get('cycle_length', 28) > 38:
            alerts.append(AnomalyAlert(
                anomaly_type=AnomalyType.LONG_CYCLE,
                severity='medium',
                description=f"Ciclo largo ({record['cycle_length']} d√≠as).",
                recommendation="Puede indicar anovulaci√≥n. Monitorea pr√≥ximos ciclos.",
                confidence=0.90
            ))
        
        if record.get('unusual_bleeding', 0) == 1:
            alerts.append(AnomalyAlert(
                anomaly_type=AnomalyType.UNUSUAL_BLEEDING,
                severity='high',
                description="Sangrado inusual fuera del per√≠odo.",
                recommendation="Consulta m√©dica si persiste.",
                confidence=0.85
            ))
        
        if record.get('cycle_cv', 0) > 0.15:
            alerts.append(AnomalyAlert(
                anomaly_type=AnomalyType.IRREGULAR_CYCLE,
                severity='medium',
                description=f"Alta variabilidad (CV={record['cycle_cv']:.2f}).",
                recommendation="Considera factores: estr√©s, peso, ejercicio.",
                confidence=0.80
            ))
        
        return alerts
    
    def get_summary(self, df: pd.DataFrame) -> pd.DataFrame:
        counts = {
            'short_luteal': (df['luteal_phase'] < 10).sum(),
            'long_cycle': (df['cycle_length'] > 38).sum(),
            'short_cycle': (df['cycle_length'] < 23).sum(),
            'unusual_bleeding': df['unusual_bleeding'].sum() if 'unusual_bleeding' in df.columns else 0,
            'high_variability': (df['cycle_cv'] > 0.15).sum() if 'cycle_cv' in df.columns else 0
        }
        return pd.DataFrame([{'type': k, 'count': v, 'pct': round(v/len(df)*100, 2)} for k, v in counts.items()])
````

### 5.7 Adaptive Predictor (Main System)

````python
class BloomAdaptivePredictor:
    
    def __init__(self):
        self.cycle_predictor = CycleLengthPredictor()
        self.ovulation_classifier = OvulationClassifier()
        self.anomaly_detector = AnomalyDetector()
        self.feature_engineer = BloomFeatureEngineer()
        self.user_history: Dict[int, List[Dict]] = {}
        self.is_fitted = False

    def fit(self, df: pd.DataFrame) -> 'BloomAdaptivePredictor':
        df_feat = self.feature_engineer.transform(df)
        feature_sets = self.feature_engineer.get_feature_sets()
        
        cycle_feats = [f for f in feature_sets['cycle'] if f in df_feat.columns]
        self.cycle_predictor.fit(df_feat[cycle_feats], df_feat['cycle_length'], cycle_feats)
        
        ov_feats = [f for f in feature_sets['ovulation'] if f in df_feat.columns]
        self.ovulation_classifier.fit(df_feat[ov_feats], df_feat['ovulation_day'], ov_feats)

        anom_feats = [f for f in feature_sets['anomaly'] if f in df_feat.columns]
        self.anomaly_detector.fit(df_feat, anom_feats)
        
        self.is_fitted = True
        return self
    
    def add_user_cycle(self, user_id: int, data: Dict) -> None:
        if user_id not in self.user_history:
            self.user_history[user_id] = []
        self.user_history[user_id].append(data)
    
    def predict(self, user_id: int, features: pd.Series) -> BloomPrediction:
        features_df = pd.DataFrame([features])
        
        cycle_pred, lower, upper = self.cycle_predictor.predict_with_interval(features_df)
        base_cycle = cycle_pred[0]
        
        ovulation_pred = self.ovulation_classifier.predict(features_df)
        anomalies = self.anomaly_detector.detect(features)
        
        if user_id in self.user_history and len(self.user_history[user_id]) >= 3:
            final_cycle, source = self._adapt(user_id, base_cycle)
        else:
            final_cycle, source = base_cycle, 'model_only'
        
        return BloomPrediction(
            user_id=user_id,
            predicted_cycle_length=round(final_cycle, 1),
            cycle_confidence_interval=(round(lower[0], 1), round(upper[0], 1)),
            ovulation=ovulation_pred,
            fertile_window_start=ovulation_pred.window_start - 3,
            fertile_window_end=ovulation_pred.window_end,
            anomaly_alerts=anomalies,
            prediction_source=source
        )
    
    def _adapt(self, user_id: int, model_pred: float) -> Tuple[float, str]:
        cycles = [c['cycle_length'] for c in self.user_history[user_id]]
        mean, std = np.mean(cycles), np.std(cycles)
        
        weight = 0.7 if std < 2 else (0.5 if std < 3.5 else 0.3)
        
        trend = 0
        if len(cycles) >= 4:
            trend = np.mean(cycles[-2:]) - np.mean(cycles[-4:-2])
        
        adapted = weight * (mean + trend * 0.2) + (1 - weight) * model_pred
        return adapted, f'adaptive (w={weight})'
    
    def get_user_stats(self, user_id: int) -> Optional[Dict]:
        if user_id not in self.user_history:
            return None
        cycles = [c['cycle_length'] for c in self.user_history[user_id]]
        return {
            'n_cycles': len(cycles),
            'mean': round(np.mean(cycles), 1),
            'std': round(np.std(cycles), 1),
            'regularity': 'regular' if np.std(cycles) < 3 else 'irregular'
        }
````
